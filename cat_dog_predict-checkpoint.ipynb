```python
cat_dir = os.path.join('data/Cat')
dog_dir = os.path.join('data/Dog')

catNames = os.listdir(cat_dir) 
dogNames = os.listdir(dog_dir)

fileTargetIndex = 209

cat_images = [os.path.join(cat_dir, fname)
              for fname in catNames[fileTargetIndex-8: fileTargetIndex]]
dog_images = [os.path.join(dog_dir, fname)
              for fname in dogNames[fileTargetIndex-8: fileTargetIndex]]

for i, img_path in enumerate(cat_images + dog_images):
    sp = plt.subplot(4, 4, i+1)
    sp.axis('off')
    img = cv2.imread(img_path) # robust library for image processing and can handle a wider range of image formats
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)

    plt.title(f"Image {i+1}")  # Add a title to each image
    
    # Print dimensions of the current image
    print(f"Image {i+1} dimensions: {img.shape}")

plt.show()
```

```python
base_dir = 'data'

# creating datasets
train_data = image_dataset_from_directory(
    base_dir,
    image_size=(200, 200),
    subset='training',
    seed=1,
    validation_split=0.1,
    batch_size=32
)
test_datagen = image_dataset_from_directory(
    base_dir, 
    image_size=(200,200), 
    subset='validation', 
    seed = 1, 
    validation_split=0.1, 
    batch_size= 32
)

print("Train data element specification:")
print(train_data.element_spec)

print("\nTest data element specification:")
print(test_datagen.element_spec)
```

```python
model = tf.keras.models.Sequential([
  layers.Conv2D(32, (3, 3), activation='relu', input_shape=(200, 200, 3)),
  layers.MaxPooling2D((2, 2)),
  layers.Conv2D(64, (3, 3), activation='relu'),
  layers.MaxPooling2D((2, 2)),
  layers.Conv2D(64, (3, 3), activation='relu'),
  layers.MaxPooling2D((2, 2)),
  layers.Conv2D(64, (3, 3), activation='relu'),
  layers.MaxPooling2D((2, 2)),
  layers.Flatten(),
  layers.Dense(512, activation='relu'),
  layers.BatchNormalization(),
  layers.Dense(512, activation='relu'),
  layers.Dropout(0.1),
  layers.BatchNormalization(),
  layers.Dense(512, activation='relu'),
  layers.Dropout(0.2),
  layers.BatchNormalization(),
  layers.Dense(1, activation='sigmoid')
])

try:
  model.summary()
except Exception as e:
  print( e)
```
